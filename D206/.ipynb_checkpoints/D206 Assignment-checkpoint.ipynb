### C1. Detecting Data Quality Issues – Discussion

**Duplicates -** Duplicates will be detected using the pandas built-in method, df.duplicated.

**Missing Values -** Missing Values will be detected using df.info() and df.isna().sum().

**Outliers –** Outliers will be detected using df.column.std() to calculate the z-score of the datapoints thus identifying which values are considered outliers.

**Re-expression of categorical variables –** Columns which are expressed inaccurately and have the wrong data type will be detected using the df.info() method.

<br>
<a id='c2'></a>

### C2. Detecting Data Quality Issues – Justification

**Duplicates –** The df.duplicated method returns a True or False value for each row if duplicates are found in a given a subset of data. The returned value can later be used to filter the dataframe to only show non-duplicate data.

**Missing Values –** The df.info() method returns a Non-Null count for all columns in the data whilst df.isna().sum() counts null values per column.

**Outliers –** df.column.std() to calculate the z-score of the datapoints thus identify which values are considered outliers. An outlier is considered any data point with an absolute z-score of greater than 3 – *wgu courseware lesson 6, Z-Scores.*

**Re-expression of categorical variables –** The df.info() method shows the datatypes of each column in the dataset. The datatype can later be re-expressed after reviewing the data dictionary to determine which columns should be considered categorical.

### C4. Detection Code – *See code below*

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 100)
```

```python
# Read csv file
df = pd.read_csv('churn_raw_data.csv', index_col=0)
```

```python
# Duplicates 
df[df.duplicated(subset = ['City', 'State', 'County' , 'Zip', 'Lat', 'Lng', 'Gender','Population'] ,keep = 'last')]

# Missing Values
df.info()
df.isna().sum()

# Outliers
df[abs(df.Population - df.Population.mean())/df.Population.std() > 3] 
df[abs(df.Children - df.Children.mean())/df.Children.std() > 3]
df[abs(df.Age - df.Age.mean())/df.Age.std() > 3]
df[abs(df.Income - df.Income.mean())/df.Income.std() > 3]
df[abs(df.Outage_sec_perweek - df.Outage_sec_perweek.mean())/df.Outage_sec_perweek.std() > 3]
df[abs(df.Email - df.Email.mean())/df.Email.std() > 3]
df[abs(df.Contacts - df.Contacts.mean())/df.Contacts.std() > 3]
df[abs(df.Yearly_equip_failure - df.Yearly_equip_failure.mean())/df.Yearly_equip_failure.std() > 3]
df[abs(df.Tenure - df.Tenure.mean())/df.Tenure.std() > 3]
df[abs(df.Bandwidth_GB_Year - df.Bandwidth_GB_Year.mean())/df.Bandwidth_GB_Year.std() > 3]

numerics = ['Population', 'Children', 'Age', 'Income', 'Outage_sec_perweek',
       'Email', 'Contacts', 'Yearly_equip_failure', 'Tenure', 'MonthlyCharge',
       'Bandwidth_GB_Year']

sns.set_style('dark')
fig, axes = plt.subplots(nrows=6, ncols=2, figsize=(15, 20))
for i in range(11):
    df.boxplot(numerics[i], ax=axes[i//2, i%2])

#Re-expression of categorical variables 
df.info()
```

###### Duplicates

```python
# Fitlering the dataset for non-duplicate rows
df = df[~df.duplicated(
    subset = ['City', 'State', 'County' , 'Zip', 'Lat', 'Lng', 'Gender','Population'] ,
    keep = 'last')]
```

Duplicates where checked for a subset of the data which included: City, State, County , Zip, Latitude, Longitude and Gender. The assumption is that whilst fields like age, marital status, prefrences and job may have changed over time, a person living at the same address will most likely be the same person duplicated. After detection the dataframe was sliced to only keep the last row of the duplicate, with the assumption that the last row is the most up to date.

```python
# filling missing Age values with the mean
df.Age.fillna(df.Age.mean(), inplace=True)
```

The second column with missing values to be treated is the age column. All missing values were replaced with the mean as it seems the most effective solution.

```python
# filling missing Income values with the median
df.Income.fillna(df.Income.median(), inplace=True)
```

The third column with missing values to be treated was the income column. All missing values were replaced with the median. The reasoning being that the data is right skewed and using the mean wouldn't be an accurate presentation of the data.

```python
# get the mode for the techie column
techie_mode = df.Techie.mode().item()
# filling missing Techie values with the mode
df.Techie.replace(np.nan, techie_mode, inplace=True)
```

The fourth column with missing values to be treated was the Techie column. All missing values were replaced with the mode. The reasoning being that the data is categorical, either yes or no, and do not have an average.

The fifth column with missing values to be treated was the Phone column. All missing values were replaced with the mode. The reasoning being that the data is categorical, either yes or no, and do not have an average.

```python
# get the mode for the TechSupport column
tech_support_mode = df.TechSupport.mode().item()
# filling missing TechSupport values with the mode
df.TechSupport.replace(np.nan, tech_support_mode, inplace=True)
```

The sixth column with missing values to be treated was the TechSupport column. All missing values were replaced with the mode. The reasoning being that the data is categorical, either yes or no, and do not have an average.

### D6. Disadvantages<a id='d6'></a>
Duplicates were searched for a subset of the data that can be summarized as address and gender. However just because two customers have the same address doesn’t provide sufficient evidence that there are duplicate records. Customers may live together, or a new customer moved in after the previous customer moved out. The address may also be that of a building, with multiple apartments. A possible effective approach would be to check the age ranges, and were there a way to ascertain the input date of each record, to determine if the ages match, and if not, then not to consider the records as duplicate.

Rows with missing values for children could have been filled with the mean instead of the mode, and the chosen way may overfit the chosen method of analysis. 

Even though the outliers in the dataset where determined to be due to the expected spread in the data and not due to error, the outliers may have impact on the chosen method of analysis and may impact the results resulting in either over or under fitting.

The children column can have been re-expressed as a category, being that the max children in the dataset is 10. Nonetheless, being that the amount of children a person may have is theoretically a continuous variable, the children variable was left as a integer.

### D7. Challenges <a id='d7'></a>
Due to the lack of data and the assumption that a duplicate customer is one that has the same address, the results of the analysis may be flawed. If for example a customer had a job of a surgeon and did not churn, then moved and still remained a customer, we may incorrectly assume customers working in the medical field, especially the surgeons, are less likely to churn.

Another possible issue an analyst may encounter, the model mmay be overfit due to the choice of replacing missing values for children with the mode. The mean would lead to less bias. 

### E1. PCA <a id='e1'></a>

```python
# The following code domes from the datacamp course
# [Dimensionality Reduction in Python, Part 4, Calculating Principal Components] 
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Scale the data
scaler = StandardScaler()
df_std = scaler.fit_transform(df[numerics])

# Apply PCA
pca = PCA()
pc = pca.fit_transform(df_std)
```

### E2. Principal Components<a id='e2'></a>

```python
# Scree Plot [WGU Courseware]
variance = pca.explained_variance_ratio_
plt.plot(variance)
plt.xlabel('number of components')
plt.ylabel('explained variance')
plt.show()
```

The method used to determine the principal components was to use a scree plot. Viewing the above plot, we can determine the elobow of the plot sits at the index number 1, in others words PC2. Thus the pricipal components that will be kept are PC1 and PC2.
